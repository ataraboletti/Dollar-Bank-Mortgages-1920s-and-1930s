This data set was prepared by Adrianna Taraboletti, a graduate student enrolled at the University of Pittsburgh's Master's in Library and Information Science program. The data was collected from five mortgage books from the Dollar Bank Archive Collection. Dollar Bank is a mutual savings bank founded in 1855, making it Pittsburgh's oldest bank. The data includes information about mortgages approved by Dollar Bank in the 1920s and 1930s. For each mortgage the names of mortgage holders, the year the mortgages were approved, the principal amount of the mortgages, the interest rates of the mortgages when they were approved, and the foreclosure date of mortgages that were foreclosed on were collected. This data is published with the permission of Robert Presutti, the archivist for Dollar Bank.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### The Data

```python
mortgage_data = pd.read_csv("Files/Dollar Bank Pittsburgh Mortgages_1920s-1930s.csv")
mortgage_df = pd.DataFrame(mortgage_data)
mortgage_df
```

```python
selected_mortgage_data = mortgage_df.drop("Foreclosure Date (Year)", axis=1)
selected_mortgage_data
```

```python
foreclosure_data = mortgage_df.dropna()
foreclosure_data = foreclosure_data.astype({"Foreclosure Date (Year)" : int})
foreclosure_data
```

### Number of Mortgages Approved Per Year

```python
results = []

for year in selected_mortgage_data.groupby('Date (Year)'):
    x = year[0]
    y = year[1]['Date (Year)'].count()
    one_set_of_results=[x,y]
    results.append(one_set_of_results)
mortgage_years_totals = pd.DataFrame(results)
mortgage_years_totals.columns=['Date (Year)', 'Number of Mortgages']
mortgage_years_totals
```

```python
mortgage_years_totals.plot.bar(x='Date (Year)', y='Number of Mortgages',
                                figsize=(8,8), title='Number of Mortgages Approved Per Year')
plt.show()
```

It can be seen by looking at the above graph that there is a significant drop-off of approved mortgages between 1931 and 1932. This can be attributed to the economic crisis that would be known as the Great Depression. The Great Depression began as a result of the stock market crash of October 1929. Thus, it makes sense that there would be a massive drop-off a few years after this event because it would have taken some time for Americans to realize that the economic crisis would span multiple years rather than being a short-term crisis.

### Average Principal Amount of Mortgages Per Year

```python
results_1 = []

for year in selected_mortgage_data.groupby('Date (Year)'):
    x = year[0]
    y = year[1]['Principal Amount'].mean()
    one_set_of_results_1=[x,y]
    results_1.append(one_set_of_results_1)
mortgage_principal_averages = pd.DataFrame(results_1)
mortgage_principal_averages.columns=['Date (Year)', 'Average Principal Amount']
mortgage_principal_averages
```

```python
mortgage_principal_averages.plot.bar(x='Date (Year)', y='Average Principal Amount',
                                figsize=(8,8), title='Average Principal Amount of Mortgages Per Year')
plt.show()
```

From the graph above, it can be seen that the average principal value of approved mortgages dipped in the mid-years of the 1930s, most likely as a result of the Great Depression. However, it is interesting to see that the values increased slightly (with the exception of 1938), showing the economic recovery that was occurring in the late-1930s as a result of the passing of the New Deal.

### Average Interest Rate Per Year

```python
results_2 = []

for year in selected_mortgage_data.groupby('Date (Year)'):
    x = year[0]
    y = year[1]['Interest Rate (Percent)'].mean()
    one_set_of_results_2=[x,y]
    results_2.append(one_set_of_results_2)
mortgage_interest_averages = pd.DataFrame(results_2)
mortgage_interest_averages.columns=['Date (Year)', 'Average Interest Rate']
mortgage_interest_averages
```

```python
mortgage_interest_averages.plot.bar(x='Date (Year)', y='Average Interest Rate',
                                figsize=(8,8), title='Average Interest of Mortgages Per Year')
plt.show()
```

From the above graph, it can be seen that the average interests rates did not change significantly until 1934. This can be attributed to the banking reforms that were passed by the federal government as part of the New Deal. The first Banking Act was passed in 1933 and a second act was passed in 1935.
